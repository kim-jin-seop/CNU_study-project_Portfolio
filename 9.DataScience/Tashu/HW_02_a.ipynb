```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
```

```python
train_data = pd.read_csv('./Tashu/total_rent.csv')
train_data['MONTH'] = train_data['MONTH'].astype('object')
train_data['HOUR'] = train_data['HOUR'].astype('object')
train_data['WEEKDAY'] = train_data['WEEKDAY'].astype('object')
train_data['SEASON'] = train_data['SEASON'].astype('object')
train_data.head(5)
```

```python
train_data = train_data.fillna(0)
del train_data['STATION']
del train_data['YEAR']
del train_data['DAY']
train_data.head()
```

```python
X_train = train_data.iloc[:, :-1].values
y_train = train_data.iloc[:, -1].values
```

```python
# Randomforest를 통한 학습 진행 tree 개수는 50개 
rf = RandomForestRegressor(n_estimators= 50)
rf.fit(X_train, y_train)
```

```python
Importance = np.rou nd(rf.feature_importances_,2)
print(Importance)
```

```python
TrainColume = ('TEMPERATURE','RAINFALL','WINDSPEED','HUMIDITY','SNOWFALL','MONTH','HOUR','WEEKDAY','SEASON')
df ={'Imt' : Importance, 'Colume' : TrainColume}
df = pd.DataFrame(df)
print(df)
```

```python
ax = sns.barplot(x=df['Imt'], y=df['Colume'])
ax.set(xlabel='Importance', ylabel='Variable')
plt.title('The number of rented bike by weekday')
plt.show()
```
